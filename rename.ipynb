```python
import os
import glob
import shutil

pf = '__5__'
for filepath in glob.glob(f'result/vv5.{pf}.*.*.*'):
    print()
    print(filepath)
    expname = filepath.split(pf+'.')[1].split('.')[0]
    print(expname)
    fname = filepath.split(expname+'.')[1].split('.')[0]
    print(fname)
    ra = filepath.split(fname+'.')[1].split('__.')[0]+'__'
    print(ra)
    ext = filepath.split(ra)[1]
    print(ext)

    model_name = fname.split('_')[0]
    ftype = fname.split('_')[1]
    new_fpath = f'result/vv5.{pf}.{model_name}.{expname}.{ftype}.{ra}{ext}'
    print(new_fpath)
    # new_fpath = filepath.replace('vv2-5', 'vv5')
    # print(filepath, new_fpath)
    os.rename(filepath, new_fpath)
    # shutil.copy(filepath, new_fpath)
```

```python
import os
import glob
import shutil

for filepath in glob.glob('result/vv5.__1__.*'):
    print()
    prepart = filepath.split('__1__')[1].split('__')[0]
    print('prepart', prepart)
    ftype = prepart.split('.')[-2]
    print('ftype', ftype)
    firstpart = prepart.split(ftype)[0]
    print('firstpart', firstpart)
    postpart = filepath.split(prepart)[1]
    print('postpart', postpart)
    roc = postpart.split('__')[1]
    print('roc', roc)
    endpart = postpart.split(roc+'__')[1]
    print('endpart', endpart)
    new_roc = '{:.2f}'.format(float(roc))
    print('new_roc', new_roc)
    new_fpath = f'result/vv5.__1__{firstpart}__{new_roc}__.{ftype}{endpart}'
    # new_fpath = filepath.replace('detected', '__81.91__.detected')
    print('new_fpath', new_fpath)
    os.rename(filepath, new_fpath)
    # shutil.copy(filepath, new_fpath)
```
